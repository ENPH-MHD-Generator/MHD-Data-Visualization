# Import and Plot Mesh

```python
import numpy as np
import pyvista as pv
import trame

pv.set_jupyter_backend("trame")
```

```python
data = pv.read("./example_data/linearHall_6.vtk")
data.plot()
```

```python
pl = pv.Plotter()
pl.add_mesh(data, show_edges=True, color='white')
pl.add_points(data.points, color='red', point_size=5)

pl.show()
```

# Colour Mesh by Scalar Values

```python
data.set_active_scalars('Potential')
data.plot(cmap='coolwarm', show_edges=True)
```

# Plot Mesh Slices

```python
data.set_active_scalars('electric_field')

slices_x = data.slice_along_axis(6)

pl = pv.Plotter()
pl.add_mesh(slices_x, cmap='coolwarm', show_edges=True)

pl.show()
```

```python
slices = data.slice_orthogonal()
pl = pv.Plotter()
pl.add_mesh(slices, cmap='coolwarm', show_edges=True)

pl.show()
```

# Plot Using Threshold Values

```python
# compute abs J
data['absJ'] = np.sqrt(np.abs(data['Jx']) + np.abs(data['Jy']) + np.abs(data['Jz']))
```

```python
data.set_active_scalars('absJ')
threshold = data.threshold(value=32)

pl = pv.Plotter()
pl.add_mesh(threshold, cmap='coolwarm', show_edges=True)
pl.show()
```

# Plot Glyphs

```python
glyphs = threshold.glyph(scale='absJ', orient='J', factor=0.0005, tolerance=0.025)
pl = pv.Plotter()
pl.add_mesh(glyphs, cmap='coolwarm', show_edges=True)
pl.show()
```

# Plot Streamlines

```python
streamlines1 = data.streamlines(vectors='J', source_radius=0.03, n_points=100)
streamlines2 = data.streamlines(vectors='J', source_center=[0.16, 0.025, 0.025], source_radius=0.03, n_points=100)
streamlines3 = data.streamlines(vectors='J', source_center=[0.04, 0.025, 0.025], source_radius=0.03, n_points=100)
pl = pv.Plotter()
pl.add_mesh(streamlines1, cmap='coolwarm', show_edges=True)
pl.add_mesh(streamlines2, cmap='coolwarm', show_edges=True)
pl.add_mesh(streamlines3, cmap='coolwarm', show_edges=True)
pl.add_mesh(data, opacity=0.3, show_edges=True)
pl.show()
```
